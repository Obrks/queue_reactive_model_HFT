Simple demonstration of the use for the implementation of Order Book Simulation using The Queue Reactive Model 

Initialisation of the Order Book

```python
import QR_model as QR

p_ref = 100.0
k = 5
tick_size = 0.1

initial_state = [11,12,13,14,15,15,14,12,11,10] 
init = initial_state + [0] * (k - len(initial_state))

p_lowest = p_ref - tick_size/2 - tick_size * (k-1)
states = [QR.Queue(p_lowest + i * tick_size, init[i]) for i in range(k * 2)]

ob = QR.OrderBook(states=states, k=k, p_ref=p_ref, tick=tick_size)
```

We may plot the Order Book and see how it moves in time

```python
from QR_plot import simulate_and_show
simulate_and_show(ob, T=10, stf=10, Cbound=200, delta=1.0, H=1000)
```

We may also plot a graph to see the best bid/ask and spread in time

```python
from QR_plot import simulate_best_prices, plot_best_prices

times, best_bids, best_asks = simulate_best_prices(
    ob, T=10, stf=10, Cbound=100, delta=0.5, H=10
)

plot_best_prices(times, best_bids, best_asks)
```

Trying different parameters to get different market situations

```python
times1, best_bids1, best_asks1 = simulate_best_prices(
    ob, T=10, stf=5, Cbound=100, delta=1.0, H=40
) 

times2, best_bids2, best_asks2 = simulate_best_prices(
    ob, T=10, stf=10, Cbound=100, delta=0.5, H=10
)

times3, best_bids3, best_asks3 = simulate_best_prices(
    ob, T=10, stf=20, Cbound=100, delta=0.2, H=5
)

plot_best_prices(times1, best_bids1, best_asks1)
plot_best_prices(times2, best_bids2, best_asks2)
plot_best_prices(times3, best_bids3, best_asks3)
```
